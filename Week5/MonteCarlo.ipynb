```python
f = lambda x:x**(0.5) + sin(x)
```

```python
x = arange(0.,15.,0.001)
```

```python
myplot = plot(x,f(x),lw=3)
ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
xlabel(r'$x$',fontsize=20)
xlim(0.,15.)
show()
```

```python
xmin = 1.7
xmax = 12.4

#find the maximum vlue of f within the boundary to set ymax
subx = x[logical_and((x>xmin),(x<xmax))]
ymax = max(f(subx))*1.05
ymin = 0.

#replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

#plot the box
plot([xmin, xmin],[ymin,ymin],color='k', linestyle='--') #vertical left
plot([xmax, xmax],[ymin,ymax], color='k', linestyle='--') #vertical right
plot([xmin, xmax],[ymax, ymax], color='k', linestyle='--') #horizontal top
```

```python
#random_sample?

N = 10000
samples_x = xmin + (xmax-xmin)*random_sample(N)
samples_y = ymin + (ymax-ymin)*random_sample(N)

#Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:200], samples_y[1:200],".")

#plot the box
plot([xmin, xmin],[ymin,ymin],color='k', linestyle='--') #vertical left
plot([xmax, xmax],[ymin,ymax], color='k', linestyle='--') #vertical right
plot([xmin, xmax],[ymax, ymax], color='k', linestyle='--') #horizontal top
```

```python
newmask = (samples_y< f(samples_x))
```

```python
sum(newmask)
```

```python
#Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:200], samples_y[1:200],".")

#Show the ones below the function in red
plot(samples_x[newmask[:200]],samples_y[newmask[:200]],'r.')

#plot the box
plot([xmin, xmin],[ymin,ymin],color='k', linestyle='--') #vertical left
plot([xmax, xmax],[ymin,ymax], color='k', linestyle='--') #vertical right
plot([xmin, xmax],[ymax, ymax], color='k', linestyle='--') #horizontal top
```

```python
I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
print I
```

Repeat the hit or miss method of Monte Carlo integration on the function $f(x) = x^{1/2} + \sin(x)$ from example 1 with different number of samples to arrive at a result that is within .001%

```python
Iana = lambda x: (2./3.)*x**(3./2.) - cos(x)
Iexact = Iana(xmax) - Iana(xmin)
Ierr = math.fabs(100*(Iexact-I)/Iexact)
print "I_exact =",Iexact
print "I_hitormiss =",I
print "Error on I=",Ierr,"%"
```

```python
#random_sample?

N = 1000000
samples_x = xmin + (xmax-xmin)*random_sample(N)
samples_y = ymin + (ymax-ymin)*random_sample(N)
newmask = (samples_y< f(samples_x))

#Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:], samples_y[1:],".")

#Show the ones below the function in red
plot(samples_x[newmask[:]],samples_y[newmask[:]],'r.')

#plot the box
plot([xmin, xmin],[ymin,ymin],color='k', linestyle='--') #vertical left
plot([xmax, xmax],[ymin,ymax], color='k', linestyle='--') #vertical right
plot([xmin, xmax],[ymax, ymax], color='k', linestyle='--') #horizontal top

I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
print I

Iana = lambda x: (2./3.)*x**(3./2.) - cos(x)
Iexact = Iana(xmax) - Iana(xmin)
Ierr = math.fabs(100*(Iexact-I)/Iexact)
print "I_exact =",Iexact
print "I_hitormiss =",I
print "Error on I=",Ierr,"%"
```
