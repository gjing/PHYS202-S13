```python
#number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N)

#inverse CDF
Finv = lambda z: arccos(cos(0)-2*z)
    
#sampled x values
xsamples = Finv(z)

#how do they compare to the original?
hist(xsamples,50,normed=True,label="Sampled distribution")
plot(x,f(x),label="True distribution",lw=3, color='red')

xlim(0,pi)
ylim(0,0.7)

legend(loc=0)
show()
```

```python
import numpy as np
x = arange(0,5,.01)
l = 1
f = lambda x: l*e**-(l*x)
def inverter(f,x):
    Fx = cumsum(f(x))/sum(f(x))
    z = numpy.random.random_sample(10000)
    xsamples = x[searchsorted(Fx,z)]
    return xsamples
```

```python
hist(inverter(f,x),50,normed=True)
plot(x,f(x), linewidth = 2, label = 'true distribution')
legend()
show()
```

Exercise 3: Quantum Double Slit Experiment
-

(a)

```python
x = arange(-.2, .2+.01, .001)
A = 1
a = 10E-6
l = 632.8E-9
L = 1
d = 15E-6
N = 10000
Ids = lambda x: A**4*(sinc(a*x/(l*L)))**2 * (cos(2*pi*d*x/(l*L)))**2
plot (x, Ids(x))
axis("tight")
show()
```

(b)

```python
n,b,p = hist(inverter(Ids, x), 100, normed=True, label="Sampled Distribution")
plot(x, Ids(x)*max(n), label="True Distribution", lw=3, color='red', linestyle="--", linewidth=2)
legend(loc='upper right')
show()
```
