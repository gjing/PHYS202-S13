```python
import numpy.random
```

```python
x = arange(0,30.1, 0.1)
a = 1.5
B = 6
f = lambda x: a*B**(-a)*x**(a-1)*exp(-(x/B)**a)
#a
F = lambda x: 1 - exp(-(x/B)**a)
N = 10000
z = numpy.random.random_sample(N);
#b
Finv = lambda z: B*(-log(1-z))**(1/a)
#c
xsamples = Finv(z)
#d
plot(x,f(x),label="true distribution",lw=3)
plot(x, F(x), label="CDF", lw =3)
hist(xsamples,50,normed=True,label="Sampled Distribution")
legend(loc=4)
show()
```

```python
f = lambda x: 60 * x**3 * (1 - x)**2
x = arange(0.0,1.0001,0.001)
xmin = 0
xmax = 1.0
ymax = 2.4
ymin = 0
N = 10000
samples_x = xmin + (xmax - xmin) * random_sample(N);
samples_y = ymin + (ymax - ymin) * random_sample(N);
newmask = (samples_y < f(samples_x))
myplot = plot(x,f(x),lw=3)
xlim(0.0,1.0)
plot(samples_x[1:10000],samples_y[1:10000],'.')
plot(samples_x[newmask[:10000]],samples_y[newmask[:10000]],'.')
plot([xmin, xmin], [ymin, ymax], color = 'k', linestyle = '-')
plot([xmax, xmax], [ymin, ymax], color = 'k', linestyle = '-')
plot([xmin, xmax], [ymax, ymax], color = 'k', linestyle = '-')
show()
```
