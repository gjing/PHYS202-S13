```python
def Euler(f, xa, xb, ya, n):
      dx = (xb - xa) / float(n)
      x = xa
      y = ya
      for i in range(n):
          y += dx * f(x, y)
          x += dx
      return x,y
```

```python
dydx = lambda x,y: x+2*y
Euler(dydx,0,1,0,4)
```

```python
Euler(dydx,0,1,0,50)
```

```python
a = arange(0,1.25,.25)
b = arange(0,1.02,.02)
plot(a, Euler(dydx,0,a,0,4)[1], 'r-', label="steps of .25")
plot(b, Euler(dydx,0,b,0,50)[1], 'b-', label="steps of .02")
plot(b, .25*e**(2*b) - .4*b - .25, 'g-', label="calculated result")
legend(loc = 0)
```
