```python
import scipy
import scipy.integrate as integrate
#each dx should have .1*10**-6 of charge since each dx is .1m long
f = lambda x: pointField(3,3,.1*10**-6,x,0)[0]
g = lambda x: pointField(3,3,.1*10**-6,x,0)[1]
def TrapezoidInt(a,b,f,dx):
    integral = 0
    x = a
    while x < b-dx:
        y1 = f(x)
        y2 = f(x+dx)
        y = (y1+y2)/2
        x+=dx
        integral+=y*dx
    return integral    
print TrapezoidInt(1,2,f,.1)
print TrapezoidInt(1,2,g,.1)
```

```python
def is_anagram(s1,s2):
    s3 = s1
    s4 = s2
    for let in s3:
        for ter in s4:
            if let == ter:
                s3 = s3[0:s3.index(let)] + s3[s3.index(let)+1:]
                s4 = s4[0:s4.index(ter)] + s4[s4.index(ter)+1:]
                break
    if s3 == s4:
        return True
    else:
        return False
```

```python
print is_anagram("need", "eden")
print is_anagram("bob", "ross")
```
