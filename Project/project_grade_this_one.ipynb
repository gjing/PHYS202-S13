```python
class Voxel:
    """A class to store the parameters of each voxel and calculate the gradient. Takes in a String parameter and parses it"""
    def __init__(self, data):
        self.volume = float(data.split(' ')[0])
        self.row = float(data.split(' ')[1])
        self.column = float(data.split(' ')[2])
        self.bucket = float(data.split(' ')[3])
        self.ADC = float(data.split(' ')[4])
    def setGradient(self, other):
        self.gradient = (other.ADC-self.ADC)
```

```python
def getEvent(f, e_num):
    """Parses the file, looking for the string that begins with #### Event Number
        and then creates a voxel with each subsequent string, adding it to the array
        until it hits a line that contains "####" the array is returned"""
    line = f.readline()
    atline = False
    arr = []
    while not atline:
        if "#### Event " + str(e_num) in line:
            atline = True
        line = f.readline()
    atline = False
    while not atline:
        if line == '' or "#### Event " in line:
            return arr
        else:
            arr.append(Voxel(line))
        line = f.readline()
```

```python
def sortData(data, n):
    """takes the data and an event number sorts all the voxels by highest to lowest ADC"""
    return sorted(getEvent(data,n), key=lambda Voxel: Voxel.ADC)[::-1]
```

```python
def gradCutoff(data, cutoff):
    """creates a track and pool array determined by the cutoff and returns the arrays"""
    pool = []
    v = data[0]
    track = []
    track.append(v)
    for vox in data:
        vox.setGradient(v)
        #print vox.gradient
        if vox.gradient>cutoff:
            pool.append(vox)
        else:
            track.append(vox)
            v = vox
    return track, pool
```

```python
import math
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

class Event:
    """A class to manipulate a list of Voxels"""
    def __init__(self, data):
        self.data = sorted(data, key=lambda Voxel: Voxel.ADC)[::-1]
        self.pool = []
        self.tracks = []
    def rawGraph(self):
        fig = figure()
        ax = fig.gca(projection='3d')
        eventRow = []
        eventCol = []
        eventBuc = []
        for voxel in self.data:
            eventRow.append(voxel.row)
            eventCol.append(voxel.column)
            eventBuc.append(voxel.bucket)
        p = ax.scatter(eventRow,eventCol,eventBuc)
        ax.set_xlabel('Event Row')
        ax.set_ylabel('Event Column')
        ax.set_zlabel('Event Bucket')
    def rawTracks(self, cutoff, i=0):
        """sets the tracks and orphans. It adds voxels to an array until it longer meets the gradient cutoff, then it adds that array to the tracks array
            then repeates with what is left until all the voxels are in an array in the tracks array."""
        dat = self.data
        if i<len(self.data):
            v = self.data[i]
            t = []
            for vox in dat:
                i+=1
                vox.setGradient(v)
                if vox.gradient<cutoff:
                    t.append(vox)
                else:
                    self.tracks.append(t)
                    self.rawTracks(cutoff,i)
        else:
            return None
    def setTracks(self,cutoff):
        """moves points from tracks from the tracks array of length less than 3 to the poolo"""
        self.rawTracks(cutoff)
        for track in self.tracks:
            if len(track)<3:
                for t in track:
                    self.pool.append(t)
                self.tracks.remove(track)
        self.pool = sorted(self.pool, key=lambda Voxel: Voxel.ADC)[::-1]
    def graphEvent(self, cutoff, orphans):
        self.setTracks(cutoff)
        """graphs the tracks and orphans"""
        fig = figure()
        ax = fig.gca(projection='3d')
        for t in self.tracks:
            eventR = []
            eventC = []
            eventB = []
            for vox in t:
                eventR.append(vox.row)
                eventC.append(vox.column)
                eventB.append(vox.bucket)
                p = ax.scatter(eventR,eventC,eventB, marker='^', s=10, color=numpy.random.rand(3,1))
        eventPoolRow = []
        eventPoolCol = []
        eventPoolBuc = []
        if orphans == True:
            for voxel in self.pool:
                eventPoolRow.append(voxel.row)
                eventPoolCol.append(voxel.column)
                eventPoolBuc.append(voxel.bucket)
            q = ax.scatter(eventPoolRow,eventPoolCol,eventPoolBuc, marker='v', s=10, color=numpy.random.rand(3,1))
        ax.set_xlabel('Event Row')
        ax.set_ylabel('Event Column')
        ax.set_zlabel('Event Bucket')
```
