use the pointPotential and dipolePotential to 

a) make the surface and contour plot of the total potential for the three point particles, a=-2nC charge at (0cm, .5cm) and a dipole with +/- 3 nC charges with seperation of d=1.0cm

```python
from electrostatics import *
```

```python
import numpy as np
```

```python
x,y = meshgrid(arange(-.05,.05, 0.0007),arange(-.05,.05, 0.0007))
Vdipole = dipolePotential(x,y, -3e-9, .01)
Vpt = pointPotential(x,y, 2e-9, .005,.0)
Vt = Vpt + Vdipole
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x,y,Vt,rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)
```

Using Boolean Masks to find data
=

```python
theta = arange(0.,2*pi,pi/100.)
plot(theta,sin(theta))
show()
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)
print mask
```

```python
plot(theta,sin(theta),theta[mask],sin(theta[mask]))
```

```python
y = sin(theta[mask])
jj = (y == max(y))
y[jj] = NaN
```

```python
plot(theta,sin(theta))
plot(theta[mask],y,linewidth=4)
show()
```

```python
theta = arange(0.,2*pi+.1,pi/10)
mask = (theta==0.5)
print mask
```

```python
from scipy.interpolate import *
ntheta = 0.5; newSin=0.
sinInterp = interp1d(theta,sin(theta))
```

```python
newSin = sinInterp(ntheta)
```

```python
print newSin
```

```python
plot(theta,sin(theta), ntheta,newSin,'r+',markersize=20, markeredgewidth=3)
show()
```

```python
thetaNew = arange(0,2*pi,pi/30)
sinNew = interp1d(theta,sin(theta))(thetaNew)
```

```python
plot(theta, sin(theta), thetaNew, sinNew, 'r.')
show()
```

```python
plot(thetaNew, sinNew - sin(thetaNew))
ylabel("difference between interp and exact value")
```

```python
sinNewC = interp1d(theta, sin(theta), kind = 'cubic')(thetaNew)
subplot(2,1,1)
plot(thetaNew,sinNew-sin(thetaNew),sinNewC - sin(thetaNew), 'r.')
subplot(2,2,1)
plot(thetaNew, sinNewC - sin(thetaNew), 'r.')
```
