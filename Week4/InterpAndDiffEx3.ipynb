(a) Write a function called finite Difference(x,y) that takes as arguments arrays of values x and y(x) and returns the derivative of y with respect to x. Use the centered fi
nite di
fference method so that the dydx returned by the function has the same dimensions as the x,y arrays.

```python
def finiteDifference(x,y):
    """returns the derivative of y with respect to x"""
    dydx = zeros(y.shape,float)
    dydx[1:-1] = (y[2:]-y[:-2])/(x[2:]-x[:-2])
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```

4) Numerical di
fferentiation plays an important role in fast electronic signal processing in many experimental contexts, such as the data acquisition systems of high energy particle and nuclear physics.

The data collected is often a series of fast time samples of energy deposited in a detector by a fast-moving charged particle. Due to the stochastic nature of the particle interactions and the inherent noise in the readout electronics, these signals can be very noisy. Unfortunately, numerical di

(b) Now that we do not control the spacing of the data points, we have to use higher order di
fferentiating methods to reach a smoother and more accurate derivative. Create another function, fourPtFiniteDiff(x,y),

that computes the derivative using the higher order method above. Test that it works for y = x^5 and plot the percent error from the analytical result.

```python
def fourPtFiniteDiff(x,y):
      """Returns derivative of y with respect to x with the four point method"""
      dydx = zeros(y.shape, float)
      dydx[2:-2] = (y[0:-4]-(8*y[1:-3])+(8*y[3:-1])-y[4:])/(12*(x[3]-x[2]))
      dydx[0] = (y[1]-y[0])/(x[1]-x[0])
      dydx[1] = (y[2]-y[1])/(x[2]-x[1])
      dydx[-2] = (y[-2]-y[-3])/(x[-2]-x[-3])
      dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
      return dydx
```

```python
x = arange(0,10.,0.000315)
y = x**5
z = fourPtFiniteDiff(x,y)
w = 5*(x**4)
q = z-w
title("Percent Error for 4ptdiff", fontsize = 20)
xlabel("x-value", fontsize = 16)
ylabel("Percent Error (%)", fontsize = 16)
ylim(-.000000001,.000000001)
plot(x, 100*(q/w))
```
