3) Let's explore the accuracy of numerical differentiation.
-

(a) Write a function called finite Difference(x,y) that takes as arguments arrays of values x and y(x) and returns the derivative of y with respect to x. Use the centered finite difference method so that the dydx returned by the function has the same dimensions as the x,y arrays.

```python
def finiteDifference(x,y):
    """returns the derivative of y with respect to x"""
    dydx = zeros(y.shape,float)
    dydx[1:-1] = (y[2:]-y[:-2])/(x[2:]-x[:-2])
    dydx[0] = (y[1]-y[0])/(x[1]-x[0])
    dydx[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
    return dydx
```

(b) Use your function to take the derivative of the function y = x^5 over the range 0 < x < 10, with steps of 0.1. How do you know it works correctly? (Hint: compare to the analytical result).

```python
x = arange(0,10.,0.1)
y = x**5
z = finiteDifference(x,y)
print z
```

(c) What is the accuracy of the result if you use a spacing of 0.001? (Plot the result)

```python
x = arange(0,10.,0.001)
y = x**5
z = finiteDifference(x,y)
title("Derivation", fontsize = 20)
xlabel("X", fontsize = 16)
ylabel("Difference in value", fontsize = 16)
ylim((-.001,.001))
plot(x, z - (5*(x**4)))
```

(d) For what spacing is the derivative accurate to 4 signicant figures?

```python
x = arange(0,10.,0.000315)
y = x**5
z = finiteDifference(x,y)
title("Derivation", fontsize = 20)
xlabel("X", fontsize = 16)
ylabel("Difference in value", fontsize = 16)
ylim((-.0001,.0001))
plot(x, z - (5*(x**4)))
print "Accurate to 4 sig figs with spacing approximately .000315"
```
