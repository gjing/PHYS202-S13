$V(x,y)=\frac{kq}{\sqrt{x^2 \+ (y-d)^2}}-\frac{kq}{\sqrt{x^2 \+ (y+d)^2}}$

```python
import math
import numpy as np
k=1.
q=1.
d=1.
vxy=((k*q)/((x**2.)+(y-d)**2.)**(1/2.)) - ((k*q)/((x**2.)+(y+d)**2.)**(1/2.))
```

Ex1) Make a 3D plot of the function $G(x,y)=e^{-x^2-y^2}$  over the range $-6\lt(x,y)\lt6$   using 51 bins in $x$ and 51 bins in $y$

```python
dx = 0.25
dy = 0.25
x,y = meshgrid(arange(-6.,6.,dx),arange(-6.,6.,dy))
g = exp(-x**2-y**2)

fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x,y,g,rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)
```
